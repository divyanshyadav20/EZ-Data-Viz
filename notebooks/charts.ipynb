```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt

df = pd.read_csv('../example dataset/housePricePrediction-Train.csv')
df.head()
df_num = df.copy()
```

```python
for col in df_num.columns:
    if df_num[col].dtype == np.dtype('O'):
        df_num[col] = df_num[col].astype('category')
        df_num[col] = df_num[col].cat.codes
```

```python
df = df[df.columns[:10]]
df_num = df_num[df_num.columns[:10]]
```

```python
fig, ax = plt.subplots(len(df_num.columns), figsize=(len(df_num.columns) * 2,len(df_num.columns) * 2))
for i, col in enumerate(df_num.columns):
    sns.distplot(df_num[col], hist=True, ax=ax[i])
    ax[i].set_ylabel(col, fontsize=8)
        
plt.show()
```

```python
num_plt = 0
for col in df.columns:
    if len(df[col].unique()) < 10:
        num_plt += 1

i = 0
fig, ax = plt.subplots(1, num_plt, figsize=(num_plt * 2,num_plt * 2))
for col in df.columns:
    if len(df[col].unique()) < 10:

        df[col].value_counts().plot.bar(ax=ax[i])
        i += 1

# plt.tight_layout(pad=3.0)
plt.show()
```

```python
sns.pairplot(df)
```

```python
colNames = []
for col in df.columns:
    if df[col].dtype != np.dtype('O'):
        colNames.append(col)
        
fig, ax = plt.subplots(len(colNames), figsize=(len(colNames) * 2, len(colNames) * 4))

for i, col in enumerate(colNames):
    sns.boxplot(y=df[col], ax=ax[i])
    ax[i].set_xlabel(col, fontsize=8)

plt.show()
```
